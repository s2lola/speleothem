```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_image = cv2.imread("../image/dbe/classify_1600/5.png")
img = cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img.shape
```

```python
img_dim = img.reshape((-1,3))
img_dim = np.float32(img_dim)
```

```python
img_dim.shape
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
k = 20
attempts = 10
```

```python
ret, label, center = cv2.kmeans(img_dim, k, None, criteria, attempts, cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((img.shape))
```

```python
plt.imshow(result_image)
```
